### 6.1.2 Analog Down Converter

```python
test = ugradio.pico.capture_data('50mV', dual_mode=True, divisor=1)
```

```python
# print(np.allclose(test[0], test[1]))

data = test[0] + 1j * test[1]

fft = np.fft.fftshift(np.fft.fft(data))/2**16
power = np.abs(fft)**2
freqs = np.fft.fftshift(np.fft.fftfreq(16000, 1/62.5))
print(test)
```

```python
plt.plot(test[0,0] - test[1,0])
```

```python
plt.semilogy(freqs, power[0])
plt.xlim(0,1)
plt.title('Making sure the mixer setup works')
plt.xlabel('Frequency [MHz]')
plt.ylabel('Power [arb]')
# plt.vlines(0.)
plt.show()
```

```python
np.save('Data_lab2/test.npy', test) #Confirms mixer setup works --- we see a peak at 0.5, as expected
```

```python
from tqdm import tqdm
```

```python
line_off_data = ugradio.pico.capture_data('50mV', dual_mode=True, divisor=1, nblocks=100)
avg = np.zeros(16000) # line on data i nthe upper half (named opposite by accident)
num_blocks = line_off_data.shape[1]
# for j in range(len(data)):
#     line_off_data = data[j]
for i in range(num_blocks):
    line_off_block = line_off_data[:,i,:]
    line_off = line_off_block[0] + 1j*line_off_block[1]

    fft = np.fft.fftshift(np.fft.fft(line_off))/2**16
    power = np.abs(fft)**2
    avg += power
# print("Averaged dataset " + str(j))
avg /= (num_blocks)
freqs = np.fft.fftshift(np.fft.fftfreq(16000, 1/62.5))
plt.semilogy(freqs, avg)
plt.xlim(-2, 2)
```

### 6.2.3 Two Frequency Switched Reference Measurements

```python
temp_cold = ugradio.pico.capture_data() # Take data of cold sky
temp_cold.shape
```

```python
temp_cal = ugradio.pico.capture_data() # Take data of known blackbody
temp_cal.shape
```

```python
np.save('Data_lab2/temp_cold.npy', temp_cold) 
```

```python
np.save('Data_lab2/temp_cal.npy', temp_cal)
```

```python
cold_data = temp_cold[0] + 1j*temp_cold[1]
cal_data = temp_cal[0] + 1j*temp_cal[1]

cold_data.shape, cal_data.shape
```

```python
fft_cold = np.fft.fftshift(np.fft.fft(cold_data[0]))/2**16
fft_cal = np.fft.fftshift(np.fft.fft(cal_data[0]))/2**16

power_cold = np.abs(fft_cold)**2
power_cal = np.abs(fft_cal)**2

vs = 62.5 # MHz
freqs = np.fft.fftshift(16000, 1/vs) # plots in units of MHz
```

```python
plt.figure()
fig, (ax0, ax1) = plt.subplots(1,2, figsize=(10,4), sharey=False, sharex=False, constrained_layout=True)

ax0.set_title('Power Spectrum of Cold Sky')
ax0.semilogy(freqs, power_cold, 'k')
ax0.set_ylabel('Power [arb.]')
ax0.set_xlabel('Frequency [MHz]')

ax1.set_title('Power Spectrum of Known Blackbody')
ax0.semilogy(freqs, power_cal, 'k')
ax0.set_ylabel('Power [arb.]')
ax0.set_xlabel('Frequency [MHz]')

plt.show()
```
