```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import ugradio 
import matplotlib as mpl

mpl.rcParams['axes.linewidth'] = 2 #set the value globally
plt.rcParams['mathtext.fontset'],plt.rcParams['font.family'] = 'stix', 'STIXGeneral'
plt.rcParams.update({'font.size': 22})
```

```python
import astropy.coordinates as coord
import astropy.units as u
from astropy.coordinates import SkyCoord
from matplotlib.patches import Rectangle
```

```python
xarr = [6.45]
yarr = [62.726]

FOV = (10 * (np.pi/180))/ 2 #set by the telescope. We have a 10 degree field of view

eq = SkyCoord(xarr[:], yarr[:], unit=u.deg)
gal = eq.galactic

f = plt.figure(figsize=(12,8))
ax = plt.subplot(111, projection='mollweide')
plt.grid(True)
# plt.scatter(gal.l.wrap_at('180d').radian, gal.b.radian)
ax.add_patch(Rectangle((gal.l.wrap_at('180d').radian - FOV, gal.b.radian - FOV),FOV*2 , FOV*2,
             edgecolor = 'blue',
             facecolor = 'skyblue',
             alpha=0.5,
             fill=True,
             lw=3))

plt.show()
```
