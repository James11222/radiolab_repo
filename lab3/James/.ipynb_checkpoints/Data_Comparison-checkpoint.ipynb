```python
import ugradio
from astropy.coordinates import SkyCoord
from astropy.time import Time
import time
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import mpl_toolkits.axes_grid1 as axgrid
from mpl_toolkits.axes_grid1 import make_axes_locatable
import scipy

import astropy.constants as con
import astropy.units as u

mpl.rcParams['axes.linewidth'] = 3 #set the value globally
plt.rcParams['mathtext.fontset'],plt.rcParams['font.family'] = 'stix', 'STIXGeneral'
plt.rcParams.update({'font.size': 22})
```

```python
def format_data(dataset):
    
    val = con.c / (10.5e9 * u.Hz)
    λ = val.to(u.m).value
    L, long, alt_nch = ugradio.nch.lat, ugradio.nch.lon, ugradio.nch.alt 
    
    if dataset == "sun_6hr":
        volts_raw, times_raw = np.load('data/sun_6hr.npy')
        data = np.array([times_raw,volts_raw])
        jds=ugradio.timing.julian_date(data[0])
        α,δ =ugradio.coord.sunpos(jds[0])
        δ=(np.pi/180)*δ
        LSTs = ugradio.timing.lst(jds,lon=long) #feed in an array of julian dates observed.
        h_s = LSTs-α*(np.pi/180) 
        h_s = np.linspace(h_s[0],(h_s[0]+(15*6*(np.pi/180))),len(jds)) #h_s for 6hr sun data
        
    elif dataset == "sun_1hr":
        volts_raw, times_raw = np.load('data/1hr_Sun_data.npy')
        data = np.array([times_raw,volts_raw])
        jds=ugradio.timing.julian_date(data[0])
        α,δ =ugradio.coord.sunpos(jds[0])
        δ=(np.pi/180)*δ
        LSTs = ugradio.timing.lst(jds,lon=long) #feed in an array of julian dates observed.
        h_s = LSTs-α*(np.pi/180) 
        h_s = np.linspace(h_s[0],(h_s[0]+15*(np.pi/180)),len(jds)) #h_s for 1hr sun data 
        
    elif dataset == "moon_6hr":
        volts_raw, times_raw = np.load('data/moon.npy')
        data = np.array([times_raw,volts_raw])
        jds=ugradio.timing.julian_date(data[0])
        α,δ =ugradio.coord.moonpos(jds[0])
        δ=(np.pi/180)*δ
        LSTs = ugradio.timing.lst(jds,lon=long) #feed in an array of julian dates observed.
        h_s = LSTs-α*(np.pi/180) 
        h_s = np.linspace(h_s[0],(h_s[0]+(15*6*(np.pi/180))),len(jds)) #h_s for 6hr sun data
                                       
    else:
        print("Idk what that is mate.")
                                       
    return h_s*(180/np.pi) - 360, data[1] #hour angle array & time,voltage array
```

```python
hs_sun_6hr, data_sun_6hr = format_data("sun_6hr")
hs_sun_1hr, data_sun_1hr = format_data("sun_1hr")
hs_moon_6hr, data_moon_6hr = format_data("moon_6hr")
```

```python
z = np.polyfit(hs_sun_6hr[:17000], data_sun_6hr[:17000], deg=1)
p_sun_6hr = np.poly1d(z)

z = np.polyfit(hs_sun_1hr, data_sun_1hr, deg=1)
p_sun_1hr = np.poly1d(z)

z = np.polyfit(hs_moon_6hr, data_moon_6hr, deg=1)
p_moon_6hr = np.poly1d(z)
```

```python
f,axes = plt.subplots(1,1,sharex=True, figsize=(16,6))

axes.plot(hs_sun_6hr, (data_sun_6hr - p_sun_6hr(hs_sun_6hr)) / 2e-4, alpha=0.8, color="orange", label="6 Hour Sun Observation")


axes.plot(hs_sun_1hr + 360, (data_sun_1hr - p_sun_1hr(hs_sun_1hr)) / 2e-4, alpha=0.8, color="red", label="1 Hour Sun Observation")


axes.plot(hs_moon_6hr + 360, (data_moon_6hr - p_moon_6hr(hs_moon_6hr)) / 9e-5, alpha=0.8, color="blue",label="6 Hour Moon Observation")


axes.legend(edgecolor="black", shadow=True)
axes.set_xlabel("Hour Angle [degrees]", fontsize=30)
axes.set_ylabel("Voltage [arbitrary]", fontsize=30)
axes.hlines(y=0, xmin=-90,xmax=90, color="gray", linestyle="--")
axes.set_xlim(-75,75)
axes.set_ylim(-2,2)

plt.tight_layout()
# plt.savefig("all_raw_data.pdf", dpi=300, bbox_inches="tight")
```
