```python
import ugradio
from astropy import units as u
from astropy.coordinates import SkyCoord
from astropy.time import Time
import astropy.table as Table
from datetime import datetime
from datetime import timedelta
import time
import numpy as np
import matplotlib.pyplot as plt
import scipy

%matplotlib inline
plt.rcParams['mathtext.fontset'], plt.rcParams['font.family'] = 'stix', 'STIXGeneral'
plt.rcParams.update({'font.size': 17})
```

```python
volts, times = np.load('Data/crab_12hr.npy')
```

```python
plt.figure(figsize=(8,6))
plt.plot(times, volts, 'k')
plt.title('Raw Data - 12hr Crab Nebula Observation')
plt.xlabel('Unix Time [s]')
plt.ylabel('Volts [V]')
plt.show()
```

```python
volts_fft = np.fft.fft(volts)
power_volts = np.fft.fftshift(np.abs(volts_fft))**2

dt = np.mean(np.diff(times))
freqs = np.fft.fftshift(np.fft.fftfreq(len(volts), dt))
```

```python
%matplotlib notebook

plt.figure(figsize=(8,6))
plt.semilogy(freqs, power_volts)
plt.show()
```
