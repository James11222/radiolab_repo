```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.signal as sig
import ugradio

plt.rcParams['mathtext.fontset'],plt.rcParams['font.family'] = 'stix', 'STIXGeneral'
plt.rcParams.update({'font.size': 18})
```

```python
SIZE = 100

data = np.zeros(SIZE)
data[10:90] = 1
```

```python
FFT_0p1 = ugradio.dft.dft(data)

ACS_dft = ugradio.dft.idft(np.abs(FFT_0p1[1])**2, f=FFT_0p1[0])

ACS_nft = np.fft.fftshift(np.fft.ifft(np.abs(np.fft.fft(data))**2))

ACS_np = np.correlate(data,data, mode='same')

ACS_scipy = sig.correlate(data, data, mode="same")
```

```python
plt.plot(ACS_dft[1], label="DFT")
plt.plot(ACS_nft, linestyle="--", label="NFT")
plt.plot(ACS_np, label="numpy")
plt.plot(ACS_scipy, label="Scipy")

plt.legend()
plt.show()
```
